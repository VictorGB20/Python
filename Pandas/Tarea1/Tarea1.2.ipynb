Importamos librerias

```python
import pandas as pd
import numpy as np
```

Llamamos al fichero excel

```python
df = pd.read_csv("personas_sucio.csv")
df
```

- No haya valores núlos.

```python
df = df.fillna(0)
df
```

- Todos los nombres tengan la primera letra en mayúscula y el resto minúsculas.

```python
df['nombre'] = df['nombre'].astype(str).str.strip().str.title()
df
```

- Todas las edades estén en formato númerico (las que sean 0 o NaN, sustituirlas por el valor medio de la columna EDAD).

```python
df['edad'] = pd.to_numeric(df['edad'], errors='coerce')
media_edad = df['edad'].replace(0, np.nan).mean()
df['edad'] = df['edad'].replace(0, np.nan).fillna(media_edad).astype(int)
df
```

- Todas las ciudades estén escritas en mayúsculas.

```python
df['ciudad'] = df['ciudad'].astype(str).str.strip().str.upper()
df
```

- Todos los emails tengan el formato correcto (texto@otrotexto.com)

```python
df['email'] = df['email'].astype(str).str.strip()
patron_email = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
df.loc[~df['email'].str.match(patron_email, na=False), 'email'] = 'desconocido@correo.com'
df
```

- Todas las fechas estén en formato DD/MM/AAAA.

```python
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'], errors='coerce', dayfirst=True)
df['fecha_registro'] = df['fecha_registro'].fillna(pd.Timestamp.now())
df['fecha_registro'] = df['fecha_registro'].dt.strftime('%d/%m/%Y')
df
```

- Creamos nuevo archivo CSV

```python
df.to_csv("datos_limpios.csv", index=False)
```
